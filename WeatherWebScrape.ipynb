```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import csv
import time

# Set up the Selenium WebDriver

driver = webdriver.Chrome()

# Fetch the page
driver.get("https://www.wunderground.com/history/monthly/us/mn/fort-snelling/KMSP/date/2020-11")

# Wait for the table to be present
wait = WebDriverWait(driver, 9)
wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "tr.ng-star-inserted")))

# Extract the rows of the table
rows = driver.find_elements(By.CSS_SELECTOR, "table tbody tr")

# Prepare data for CSV
Nov = 30
Oct = 31
month = Nov
year = [2020]
day = []
temp = []
dew = []
humidity = []
wind = []
pressure = []
precip = []
recording = False



# Iterate through each row and extract cell data starting after "Nov" is found
for row in rows:
    cells = row.find_elements(By.CSS_SELECTOR, "tr")
    row_data = [cell.text.strip() for cell in cells]
    
    # Check if "Nov" is in the row_data to start recording
    if not recording:
        if any("Nov" in cell for cell in row_data):
            recording = True
            for a in range(1,month+1):
                day.append(row_data[a])
                
            for b in range(month+2, month*2 + 2):
                temp.append(row_data[b])
                
            for c in range(month*2 + 3, month*3 + 3):
                dew.append(row_data[c])

            for d in range(month*3 + 4, month*4 + 4):
                humidity.append(row_data[d])

            for e in range(month*4 + 5, month*5 + 5):
                wind.append(row_data[e])

            for f in range(month*5 + 6, month*6 + 6):
                pressure.append(row_data[f])

            for g in range(month*6 + 7, month*7 + 7):
                precip.append(row_data[g])
        continue
        
driver.quit()

data = list(zip(day, temp, dew, humidity, wind, pressure, precip))

# Write data to CSV file
with open("2020.csv", "w", newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    
    # Write the header row
    headers = ["Nov", "Temperature - MAX, AVG, MIN", "Dew Point - MAX, AVG, MIN", "Humidity - MAX, AVG, MIN", "Wind Speed - MAX, AVG, MIN", 
               "Pressure - MAX, AVG, MIN", "Precipitation - MAX, AVG, MIN", "Year"]
    writer.writerow(headers)

    # Write the data rows
    for i in range(month):
        writer.writerow([day[i], temp[i], dew[i], humidity[i], wind[i], pressure[i], precip[i], year[0]])

print("Data successfully written to 'scraped_weather_data.csv'")
```
